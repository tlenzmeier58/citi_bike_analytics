```python
# Import dependencies

import pandas as pd
import csv
import numpy as np
import os
```

```python
# June 2022 bike trip data
bikes_2022 = pd.read_csv("./Data/JC-202206-citibike-tripdata.csv")
bikes_2022.head()

# Get the columns we are going to use
bikes_2022 = bikes_2022[['ride_id', 'rideable_type', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', \
    'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng', 'member_casual']]
bikes_2022.head()
```

```python
# June 2021 bike trip data
bikes_2021 = pd.read_csv("./Data/JC-202106-citibike-tripdata.csv")
bikes_2021.head()

# Get the columns we are going to use.
bikes_2021 = bikes_2021[['ride_id', 'rideable_type', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', 'end_station_name', \
    'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng', 'member_casual']]

bikes_2021.head()
```

```python
# June 2020 bike trip data
bikes_2020 = pd.read_csv("./Data/JC-202006-citibike-tripdata.csv")

# Get the columns we are going to use.
bikes_2020 = bikes_2020[['start station id', 'starttime', 'stoptime', 'start station name', 'start station latitude', 'start station longitude', 'end station id', \
    'end station name', 'end station latitude', 'end station longitude', 'usertype']]

bikes_2020 = bikes_2020.rename(columns={
    'start station id':'start_station_id',
    'start station name':'start_station_name',
    'start station latitude':'start_lat',
    'start station longitude':'start_lng',
    'end station id':'end_station_id',
    'end station name':'end_station_name',
    'end station latitude':'end_lat',
    'end station longitude':'end_lng',
    'usertype':'member_casual',
    'starttime':'started_at',
    'stoptime':'ended_at'
})

# Add row_num
bikes_2020['ride_id'] = np.arange(len(bikes_2020))
bikes_2020['rideable_type'] = 'bike'

# Arrange the columns with the other csv files before concatenating.
bikes_2020 = bikes_2020[['ride_id', 'rideable_type', 'started_at', 'ended_at', 'start_station_name', 'start_station_id', \
    'end_station_name', 'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng', 'member_casual']]
bikes_2020.head()
```

```python
bikes_2022 = bikes_2022.copy()
bikes_2021 = bikes_2021.copy()
bikes_2020 = bikes_2020.copy()

bike_activity = pd.concat([bikes_2022, bikes_2021, bikes_2020], ignore_index=True)
bike_activity['ride_date'] = bike_activity['started_at']

# Drop NaN values from dataframe.
bike_activity = bike_activity.dropna()

# Push dataframe to csv file for Tableau
bike_activity.to_csv("./Data/bike_activity.csv")
```
